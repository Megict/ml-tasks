```python
import pandas as pd
```

```python
data = pd.read_csv('../input/adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
men_cnt = 0
wmn_cnt = 0
for v in data['sex'].to_dict().values():
    if v == 'Male':
        men_cnt += 1
    else:
        wmn_cnt += 1
print('men  : ' + str(men_cnt))
print('women: ' + str(wmn_cnt))
```

```python
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
sex_slice = data['sex'].to_dict()
age_slice = data['age'].to_dict()
age_sum = 0
for i in range(data.count()['age']):
    if sex_slice[i] == 'Female':
        age_sum += age_slice[i]
print(f"{age_sum / wmn_cnt:.3}")
```

```python
data[data['sex'] == 'Female']['age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
origin_slice = data['native-country'].to_dict()
german_sum = 0
for o in origin_slice:
    if origin_slice[o] == 'Germany':
        german_sum += 1
print(f"{german_sum * 100 / data.count()['native-country']:.3}%")
```

```python
100 * (data["native-country"] == "Germany").sum() / data.shape[0]
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?**

```python
print('for more than 50k:')
print(f"mean: {data[data['salary'] == '>50K']['age'].mean():.3f}")
print(f"sd:   {(((data[data['salary'] == '>50K']['age'] - data[data['salary'] == '>50K']['age'].mean())**2).mean())**0.5:.3f}")
print('for less than 50k:')
print(f"mean: {data[data['salary'] == '<=50K']['age'].mean():.3f}")
print(f"sd:   {(((data[data['salary'] == '<=50K']['age'] - data[data['salary'] == '<=50K']['age'].mean())**2).mean())**0.5:.3f}")
```

```python
data[data['salary'] == '<=50K']['age'].std()
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
for (race, sex), df in data.groupby(['race', 'sex']):
    print((race, sex))
    print(df['age'].describe())
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
# for marries
sum_a_lot = 0
sum_not_a_lot = 0
for elm in data[(data["sex"] == "Male")
                & (data["marital-status"].str.startswith("Married"))]['salary'].value_counts().to_dict().items():
    if elm[0] == '>50K':
        sum_a_lot = elm[1]
    else:
        sum_not_a_lot = elm[1]
print(f"a-lot part: {sum_a_lot / (sum_a_lot + sum_not_a_lot):.3f}")
```

```python
# for singles
sum_a_lot = 0
sum_not_a_lot = 0
for elm in data[(data["sex"] == "Male")
                & ~(data["marital-status"].str.startswith("Married"))]['salary'].value_counts().to_dict().items():
    if elm[0] == '>50K':
        sum_a_lot = elm[1]
    else:
        sum_not_a_lot = elm[1]
print(f"a-lot part: {sum_a_lot / (sum_a_lot + sum_not_a_lot):.3f}")
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
num_ = data[data['hours-per-week'] == data['hours-per-week'].max()].shape[0]
num_a_lot_ = data[(data['hours-per-week'] == data['hours-per-week'].max()) & (data['salary'] == '>50K')].shape[0]

print("number: " + str(num_))
print("part a lot: " + str(num_a_lot_ / num_))
```
